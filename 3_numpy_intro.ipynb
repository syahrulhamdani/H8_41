```python
arr = np.arange(0, 1, 5e-2)[:, None]
print(arr)
print(arr.shape)
arr = arr.reshape((10, 2))
print(arr.shape)
arr_2 = arr.dot(2)
print(arr_2)
mm = np.matmul(arr, arr_2.T)
print(mm)
print(mm.shape)
```

```python
mm_mean = mm.mean(axis=0)
print(mm_mean)
mm_median = np.median(mm, axis=0)
print(mm_median)
mm_std = mm.std(axis=0)
print(mm_std)
mm_var = mm.var(axis=0)
print(mm_var)
```
