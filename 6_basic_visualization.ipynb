```python
import numpy as np
import pandas as pd
```

```python
df_canada = pd.read_excel(
    "https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true",
    sheet_name="Canada by Citizenship",
    skiprows=range(20),
    skipfooter=2
)
```

```python
df_canada.drop(
    columns=[
        "AREA", "REG", "DEV",
        "Type", "Coverage"
    ],
    inplace=True
)
```

```python
df_canada.head()
```

```python
df_canada.rename(
    columns={
        "OdName": "Country",
        "AreaName": "Continent",
        "RegName": "Region"
    },
    inplace=True
)
```

```python
df_canada.head()
```

```python
df_canada["Total"] = df_canada.sum(axis=1)
df_canada.head(3)
```

```python
years = list(range(1980, 2014))
year80s = list(range(1980, 1990))
year90s = list(range(1990, 2000))
year00s = list(range(2000, 2010))
```

```python
haiti = df_canada.loc[
    df_canada["Country"] == "Haiti", years
].transpose()
haiti.rename(columns={75: "Haiti"}, inplace=True)
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot()

plt.ylabel("Number of Immigration from Haiti")
plt.xlabel("Years")
plt.title("Immigration from Haiti over Years")

plt.text(1998, 6000, "2010 Earthquake")

plt.show()
```

```python
cn_in = df_canada.loc[
    df_canada["Country"].isin(["China", "India"]),
    years
].transpose()
cn_in.rename(
    columns={
        36: "China",
        79: "India"
    },
    inplace=True
)
cn_in.head()
```

```python
cn_in.plot()

plt.title("Immigrant from China and India over Years in Canada")
plt.xlabel("Years")
plt.ylabel("Number of Immigrant")
plt.show()
```

```python
top5_countries = df_canada.sort_values(
    by="Total", ascending=False, axis=0
).head()
top5_countries = top5_countries.set_index("Country")
top5_countries = top5_countries[years].transpose()
```

```python
top5_countries.head()
```

```python
top5_countries.plot(figsize=(15, 9))
plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrant")
plt.xlabel("Years")

plt.show()
```

```python
top5_countries.plot(
    kind="area",
    stacked=False,
    figsize=(20, 10)
)
plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrant")
plt.xlabel("Years")
plt.show()
```

```python
top5_countries.plot(
    kind="area",
    alpha=.25,
    stacked=False,
    figsize=(20, 10)
)
plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrant")
plt.xlabel("Years")
plt.show()
```

```python
top5_countries.plot(
    kind="area",
    alpha=.25,
    stacked=True,
    figsize=(20, 10)
)
plt.title("Top 5 Immigrant Origin over Years in Canada")
plt.ylabel("Num of Immigrant")
plt.xlabel("Years")
plt.show()
```

```python
ax = top5_countries.plot(
    kind="area",
    stacked=False,
    figsize=(20, 10)
)
ax.set_title("Top 5 Immigrant Origin over Years in Canada")
ax.set_ylabel("Num of Immigrant")
ax.set_xlabel("Years")
plt.show()
```

```python
least5_countries = df_canada.sort_values(
    by="Total", ascending=True, axis=0
).head()
least5_countries = least5_countries.set_index("Country")
least5_countries = least5_countries[years].transpose()
```
