```python
df_canada = pd.read_excel(
    "https://github.com/ardhiraka/PFDS_sources/blob/master/Canada.xlsx?raw=true",
    sheet_name="Canada by Citizenship",
    skiprows=range(20),
    skipfooter=2
)
df_canada.head()
```

```python
df_canada['total'] = df_canada.sum(axis=1)
df_canada.head()
```

```python
int(
    round(
        (
            df_canada.loc[df_canada.OdName == "Andorra", "total"] / df_canada['total'].sum()
        ) * 100
    )
)
```

```python
max_words = 90
text = ""
for country in df_canada.OdName.values:
    if len(country.split()) == 1:
        repeat_num_times = int(round(df_canada.loc[df_canada['OdName']==country, "total"] / float(df_canada['total'].sum()) * max_words))
        text = text + ((country + ' ') * repeat_num_times)
```

```python
text
```

```python
immigrant_wc = wordcloud.WordCloud(background_color="white").generate(text)
plt.figure(figsize=(14, 9))
plt.imshow(immigrant_wc)
plt.axis("off")
plt.show()
```

```python
years = list(range(1980, 2014))
df_total = pd.DataFrame(df_canada[years].sum(axis=0))
df_total = df_total.reset_index()
df_total.columns = ['year', 'total']
df_total.head()
```

```python
plt.style.use("ggplot")
sns.regplot(x="year", y="total", data=df_total)
plt.title('Total Immigrants per Year (1980 - 2013)')
plt.show()
```

```python
plt.style.use("ggplot")
plt.figure(figsize=(10, 5))
sns.regplot(
    x="year", y="total", data=df_total,
    color="green", marker='+'
)
plt.title('Total Immigrants per Year (1980 - 2013)')
plt.show()
```

```python
sf_incidents = pd.read_csv("https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/Police_Department_Incidents_-_Previous_Year__2016_.csv")
```

```python
sf_incidents.shape
```

```python
sf_incidents.head()
```

```python
limit = 50
subset_sf_incidents = sf_incidents.loc[:limit]
subset_sf_incidents
```

```python
sf_lat = 37.77
sf_long = -122.42

sf_map = folium.Map(location=(sf_lat, sf_long), zoom_start=12)
sf_map
```

```python
a = [1, 2, 3]
b = [10, 20, 30]

for i, j in zip(a, b):
    print(i, j)
```

```python
incidents = folium.map.FeatureGroup()

for lat, long in zip(subset_sf_incidents.Y, subset_sf_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            location=(lat, long),
            radius=5,
            color="yellow",
            fill=True,
            fill_color="blue",
            fill_opacity=0.6
        )
    )

sf_map.add_child(incidents)
```

```python
incidents = folium.map.FeatureGroup()

for lat, long in zip(subset_sf_incidents.Y, subset_sf_incidents.X):
    incidents.add_child(
        folium.CircleMarker(
            location=(lat, long),
            radius=5,
            color="yellow",
            fill=True,
            fill_color="blue",
            fill_opacity=0.6
        )
    )

labels = sf_incidents.Category.tolist()
for lat, long, label in zip(subset_sf_incidents.Y, subset_sf_incidents.X, labels):
    folium.Marker((lat, long), popup=label).add_to(sf_map)

sf_map.add_child(incidents)
```

```python
sf_map = folium.Map((sf_lat, sf_long), zoom_start=12)
incidents = plugins.MarkerCluster().add_to(sf_map)

for lat, long, label in zip(subset_sf_incidents.Y, subset_sf_incidents.X, labels):
    folium.Marker(
        location=(lat, long),
        icon=None,
        popup=label
    ).add_to(incidents)

sf_map
```
